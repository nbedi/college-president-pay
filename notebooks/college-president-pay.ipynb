```python
import pandas as pd
pf = pd.read_csv('../data/presidents.csv', index_col='id')
pf['col_adjusted_base_comp'] = pf["base_comp"]/pf["cost_of_living"]
```

```python
from sklearn.neighbors import NearestNeighbors
from sklearn import preprocessing
import numpy as np
initial = np.genfromtxt('../data/presidents.csv', dtype=float, delimiter=',', skip_header=1, usecols=range(7, 10))
initial[:5]
```

```python
presidents = preprocessing.normalize(initial)
presidents[:5]
```

```python
nbrs = NearestNeighbors(n_neighbors=8, algorithm='ball_tree').fit(presidents)
distances, indices = nbrs.kneighbors(presidents)
distances[:5], indices[:5]
```

```python
pf["neighbors"] = indices.tolist()

pf.loc[indices[:1][0].tolist()]
```

```python
def neighbor_averages(ns, factor):
    return pf.loc[ns][[factor]].mean()[0]

pf['col_adjusted_neighbor_average'] =  pf["neighbors"].map(lambda x: neighbor_averages(ns=x, factor='col_adjusted_base_comp'))
pf['neighbor_average'] = pf["neighbors"].map(lambda x: neighbor_averages(x, factor='base_comp'))
pf.head()
```
